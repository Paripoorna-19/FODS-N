1.Null hypothesis:no of purple flowering plants should be 750(705+45)

Alternate hypothesis:no of purple flowering plants should not be 750

```python
simulations=1755 #2000-245
n=929
```

```python
import matplotlib.pyplot as plt
import numpy as np
p=0.75
```

here we will perform a random experiment and find how many purple flowering plants are grown if its success probability is 0.75. if we find the sampling can produce 750 purple bearing plants we won't reject null hypothesis


#single random experiment

```python
test_stat=np.random.binomial(929, 0.75)
test_stat
```

As you see that when we perfrom random experiment we got no of purple flowering plants as 696

observed statistic is 750
test statistic is 696

```python
random_sample=[]
for i in range(simulations):
  random_sample.append(np.random.binomial(929,0.75))
random_sample
```

```python
import pandas as pd
sample_df=pd.DataFrame(random_sample)
sample_df
```

```python
count=0
for i in random_sample:
  if i==750:
    count+=1
p=count/simulations
p
```

As you see there is 0% chance that the no of purple flowering plants are equal to 750. here p value(0%)<alpha value(50%). so we reject the null hypothesis, the no of purpleflowering plants can't be equal to 750(since they are highly significant)

**PROOF USING HISTOGRAM**

```python
sample_df.hist()
plt.xlabel("no of purple flowering plants")
plt.title("Distribution")
plt.scatter(750,0,color='red')
```

As you see here when we perform random sampling, the no of purple flowering plants is in the range(650, 735) it is not even closer to 750(red dot). here the observed statistic is not even in the tail of histogram.so we will reject the null hypothesis that is the no of purple flowering plants can't be equal to 750.

PART B

Null hypothesis(H0): pop_mean=60


Alternate hypothesis(H1):pop_mean!=60

```python
import numpy as np
import scipy.stats as st
pop_std=13.46
n=30
sample_mean=65.10
pop_mean=60
alpha=0.05
```

```python
z=(sample_mean-pop_mean)/(pop_std/np.sqrt(n))
z
```

```python
z_critical=st.norm.ppf(alpha/2)#since it is a tow tailed test
z_critical
```

```python
if abs(z)>abs(z_critical):
  print("Null hypothesis gets rejected")
  print("The population mean is not equal to 60")
else:
  print("Null hypothesis cannot be rejected")
  print("The population mean is equal to 60")
```
